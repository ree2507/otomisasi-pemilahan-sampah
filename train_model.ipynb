```python
from ultralytics import YOLO

# 1. Load model dasar (pilih yang Nano 'n' agar ringan dan cepat di RTX 2050)
model = YOLO('yolov8n.pt') 

# 2. Definisikan lokasi file data.yaml
# Ganti path di bawah ini dengan path LENGKAP ke file data.yaml yang baru kamu edit
path_to_yaml = r"D:\grinding\project\Otomisasi Pemilahan Sampah Berbasil Visi Komputer\dataset\YOLO-Waste-Detection-1\YOLO-Waste-Detection-1\data.yaml"

# 3. Mulai Training
results = model.train(
    data=path_to_yaml,
    epochs=50,      # Jumlah putaran belajar (bisa dinaikkan nanti)
    imgsz=640,      # Ukuran gambar standar
    batch=-1,       # Sesuaikan dengan memori GPU (8, 16, atau 32)
    workers=2,      # Sesuaikan dengan memori GPU (8, 16, atau 32)
    device=0,       # Memaksa menggunakan GPU (0 adalah index GPU utama)
    name='model_sampah_v1' # Nama hasil training nanti
)
```
