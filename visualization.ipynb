```python
import pandas as pd
import matplotlib.pyplot as plt

# Membaca data yang kamu upload
df = pd.read_csv('D:/grinding/project/Otomisasi Pemilahan Sampah Berbasil Visi Komputer/runs/detect/model_sampah_v15/results.csv')
df.columns = df.columns.str.strip() # Membersihkan nama kolom

fig, ax1 = plt.subplots(figsize=(12, 6))

# Plot Akurasi (mAP50) - Garis Biru Tebal
color = 'tab:blue'
ax1.set_xlabel('Epoch')
ax1.set_ylabel('Akurasi (mAP50)', color=color, fontweight='bold')
ax1.plot(df['epoch'], df['metrics/mAP50(B)'], color=color, linewidth=3, label='Akurasi (mAP50)')
ax1.tick_params(axis='y', labelcolor=color)
ax1.grid(True, alpha=0.3)

# Plot Loss (Kesalahan) - Garis Merah Putus-putus
ax2 = ax1.twinx()  
color = 'tab:red'
ax2.set_ylabel('Validation Loss', color=color, fontweight='bold')  
ax2.plot(df['epoch'], df['val/box_loss'], color=color, linestyle='--', label='Box Loss (Validasi)')
ax2.plot(df['epoch'], df['val/cls_loss'], color='orange', linestyle='--', label='Class Loss (Validasi)')
ax2.tick_params(axis='y', labelcolor=color)

plt.title('Grafik Training Model Pemilahan Sampah (Epoch 1-50)')
fig.tight_layout()
plt.show()
```
